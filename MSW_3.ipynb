## Knihovny

```python
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy.linalg import solve
from scipy.sparse.linalg import cg
```

## Výpočet

```python
size = range(1, 250, 5)

p_metoda = []
i_metoda = []
for velikost in size:
    a = np.random.rand(velikost, velikost)
    b = np.random.rand(velikost)

    start = time.time()
    p = solve(a, b)
    end = time.time()
    p_metoda.append(end - start)

    start = time.time()
    i = cg(a, b)
    end = time.time()
    i_metoda.append(end - start)

plt.plot(size, p_metoda, label='Metoda přímá')
plt.plot(size, i_metoda, label='Metoda iterační')
plt.xlabel('Velikost matice')
plt.ylabel('Čas v sekundách')
plt.legend()
plt.show()
```
